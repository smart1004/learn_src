```python
%matplotlib inline 
from __future__ import division
import numpy as np
import pylab
import evaluation as ev
import logistic_regression as logreg
import neighbors
```

```python
n = 6000
d = 20
k = 5
thetas = np.random.random((k, d))
X = np.random.normal(size=(n, d))
y = np.argmax(thetas.dot(X.T), axis=0)
X_train, y_train, X_test, y_test = ev.train_test_split(X, y)
```

```python
multilog = logreg.MultinomialLogisticRegression()
multilog.fit(X_train, y_train)
preds = multilog.predict(X_test)
print ev.accuracy(preds, y_test)
```

```python
print ev.accuracy(preds, y_test)
dummy = ev.DummyClassifier()
dummy.fit(X_train, y_train)

print ev.accuracy(ev.DummyClassifier().fit(X_train, y_train).predict(X_test), y_test)
```

```python
ev.benchmark(neighbors.KNNRegressor(30), X, y)
```

```python
ev.benchmark(logreg.LogisticRegression(iters=300, alpha=0.05, lambda_=0.01)
          , X, y)
```

```python
n = 100
d = 4
X = np.random.normal(size=(n, d))
y = X.dot(np.array([-1, 2, -3, 0.5])) > 0
```

```python
model = logreg.LogisticRegression()
```

```python
X_train, y_train, X_test, y_test = ev.train_test_split(X, y)
```

```python
model.fit(X_train, y_train)
ev.mse(model.predict(X_test), y_test)
```

```python
ev.auc(ev.roc_curve(model.predict(X_test), y_test))
```
